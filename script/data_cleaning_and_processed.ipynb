### II. Import dữ liệu

##### 1. Import dữ liệu bằng pandas

```python
gdp = pd.read_csv(r"..\data\raw\GDP.csv")
cpi = pd.read_csv(r"..\data\raw\CPI.csv")
cpi_percentage = pd.read_csv(r"..\data\raw\cpi_percentage.csv")
acreage_country = pd.read_csv(r"..\data\raw\acreage_country.csv")
female_population = pd.read_csv(r"..\data\raw\female_population.csv")
male_population = pd.read_csv(r"..\data\raw\male_population.csv")
gdp_percentage_capita = pd.read_csv(r"..\data\raw\gdp_percentage_capita.csv")
gdp_percentage_growth = pd.read_csv(r"..\data\raw\gdp_percentage_growth.csv")
population_country = pd.read_csv(r"..\data\raw\population_country.csv")
population_density = pd.read_csv(r"..\data\raw\population_density.csv")
unemployment_female_percentage = pd.read_csv(r"..\data\raw\unemployment_female_percentage.csv")
unemployment_male_percentage = pd.read_csv(r"..\data\raw\unemployment_male_percentage.csv")
unemployment_total_percentage = pd.read_csv(r"..\data\raw\unemployment_total_percentage.csv")
unemployment_young_female_percentage = pd.read_csv(r"..\data\raw\unemployment_young_female_percentage.csv")
unemployment_young_male_percentage = pd.read_csv(r"..\data\raw\unemployment_young_male_percentage.csv")
unemployment_young_percentage = pd.read_csv(r"..\data\raw\unemployment_young_percentage.csv")
urban_population = pd.read_csv(r"..\data\raw\urban_population.csv")
dim_country = pd.read_csv(r"..\data\cleaned\dim_country.csv")
```

### III. Làm sạch

##### 1. Xóa những vùng không cần thiết

```python
# Set id các quốc gia từ 1 file chứa id và tên đúng của các quốc gia
valid_countries = set(dim_country['id'])
```

```python
# Xóa những vùng quốc gia không cần thiết
gdp = gdp[gdp['REF_AREA'].isin(valid_countries)]
cpi = cpi[cpi['REF_AREA'].isin(valid_countries)]
cpi_percentage = cpi_percentage[cpi_percentage['REF_AREA'].isin(valid_countries)]
acreage_country = acreage_country[acreage_country['REF_AREA'].isin(valid_countries)]
female_population = female_population[female_population['REF_AREA'].isin(valid_countries)]
male_population = male_population[male_population['REF_AREA'].isin(valid_countries)]
gdp_percentage_capita = gdp_percentage_capita[gdp_percentage_capita['REF_AREA'].isin(valid_countries)]
gdp_percentage_growth = gdp_percentage_growth[gdp_percentage_growth['REF_AREA'].isin(valid_countries)]
population_country = population_country[population_country['REF_AREA'].isin(valid_countries)]
population_density = population_density[population_density['REF_AREA'].isin(valid_countries)]
unemployment_female_percentage = unemployment_female_percentage[unemployment_female_percentage['REF_AREA'].isin(valid_countries)]
unemployment_male_percentage = unemployment_male_percentage[unemployment_male_percentage['REF_AREA'].isin(valid_countries)]
unemployment_total_percentage  = unemployment_total_percentage [unemployment_total_percentage ['REF_AREA'].isin(valid_countries)]
unemployment_young_female_percentage = unemployment_young_female_percentage[unemployment_young_female_percentage['REF_AREA'].isin(valid_countries)]
unemployment_young_male_percentage = unemployment_young_male_percentage[unemployment_young_male_percentage['REF_AREA'].isin(valid_countries)]
unemployment_young_percentage = unemployment_young_percentage[unemployment_young_percentage['REF_AREA'].isin(valid_countries)]
urban_population = urban_population[urban_population['REF_AREA'].isin(valid_countries)]
```

##### 2. Đổi tên các cột 

```python
#Tạo list các dataframe
list_df = [gdp, cpi, cpi_percentage, acreage_country, female_population, male_population, gdp_percentage_capita,
            gdp_percentage_growth, population_country, population_density, unemployment_female_percentage, 
            unemployment_male_percentage, unemployment_total_percentage, unemployment_young_female_percentage,
            unemployment_young_male_percentage, unemployment_young_percentage, urban_population]
```

```python
#Set tên muốn đổi
rename_dict = {
    'REF_AREA': 'Id',
    'REF_AREA_LABEL': 'Country',
    'TIME_PERIOD': 'Year',
    'OBS_VALUE': 'Value'
                }
for df in list_df:
    df.rename(columns=rename_dict, inplace=True)
```

```python
dfs = {
    'GDP': gdp,
    'CPI': cpi,
    'CPI_PERCENTAGE': cpi_percentage,
    'ACREAGE_COUNTRY': acreage_country,
    'FEMALE_POPULATION': female_population,
    'MALE_POPULATION': male_population,
    'GDP_PERCENTAGE_CAPITA': gdp_percentage_capita,
    'GDP_PERCENTAGE_GROWTH': gdp_percentage_growth,
    'POPULATION_COUNTRY': population_country,
    'POPULATION_DENSITY': population_density,
    'UNEMPLOYMENT_FEMALE_PERCENTAGE': unemployment_female_percentage,
    'UNEMPLOYMENT_MALE_PERCENTAGE': unemployment_male_percentage,
    'UNEMPLOYMENT_TOTAL_PERCENTAGE': unemployment_total_percentage,
    'UNEMPLOYMENT_YOUNG_FEMALE_PERCENTAGE': unemployment_young_female_percentage,
    'UNEMPLOYMENT_YOUNG_MALE_PERCENTAGE': unemployment_young_male_percentage,
    'UNEMPLOYMENT_YOUNG_PERCENTAGE': unemployment_young_percentage,
    'URBAN_POPULATION': urban_population
}
```

```python
for col_name, df in dfs.items():
    df.rename(columns={'Value': col_name}, inplace=True)
```

```python
df.to_csv(r"../data/cleaned/cleaned.csv", index=False)
```

#### Merge data

```python
dfs_to_merge = [
    gdp,
    cpi,
    cpi_percentage,
    acreage_country,
    female_population,
    male_population,
    gdp_percentage_capita,
    gdp_percentage_growth,
    population_country,
    population_density,
    unemployment_female_percentage,
    unemployment_male_percentage,
    unemployment_total_percentage,
    unemployment_young_female_percentage,
    unemployment_young_male_percentage,
    unemployment_young_percentage,
    urban_population
]
```

```python
from functools import reduce
import pandas as pd

df_all = reduce(
    lambda left, right: pd.merge(
        left,
        right,
        on=['Id', 'Country', 'Year'],
        how='outer'
    ),
    dfs_to_merge
)
```
